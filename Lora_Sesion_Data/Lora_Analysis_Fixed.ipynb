# LoRa Session Data Analysis - Fixed Version
Dieses Notebook analysiert alle CSV-Dateien im Ordner und erstellt Diagramme für RSSI und SNR über die Zeit mit Farbcodierung nach Spreading Factor.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from glob import glob
import os
import numpy as np
from datetime import datetime

# Konfiguration für bessere Darstellung
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 100

# Pfad zu den CSV-Dateien
path = r'C:\Users\Admin01\Documents\Bachlore\gatewaylistener\Lora_Sesion_Data\*.csv'
files = glob(path)

print(f'Anzahl gefundener CSV-Dateien: {len(files)}\n')

# Durchlaufe alle Dateien
for file in files:
    try:
        data = pd.read_csv(file)
        file_name = os.path.basename(file)
        
        # Spalten überprüfen
        print(f'\nDatei: {file_name}')
        print(f'Anzahl Zeilen: {len(data)}')
        print(f'Spalten: {data.columns.tolist()}')

        # Prüfe ob die benötigten Spalten vorhanden sind
        if 'timestamp' in data.columns and ('rssi' in data.columns or 'rssi_dbm' in data.columns) and ('snr' in data.columns or 'snr_db' in data.columns) and 'spreading_factor' in data.columns:
            # Konvertiere timestamp zu datetime
            data['timestamp'] = pd.to_datetime(data['timestamp'])
            
            # Sortiere nach Zeit
            data = data.sort_values('timestamp')
            
            rssi_col = 'rssi' if 'rssi' in data.columns else 'rssi_dbm'
            snr_col = 'snr' if 'snr' in data.columns else 'snr_db'
            
            # Erstelle Subplots für bessere Darstellung
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10), sharex=True)
            
            # Extrahiere numerischen SF-Wert (z.B. 'SF7' -> 7)
            data['sf_numeric'] = data['spreading_factor'].str.extract('(\d+)').astype(int)
            unique_sfs = sorted(data['sf_numeric'].unique())
            
            # Verwende vielfältigere Farben für bessere Unterscheidbarkeit
            n_colors = len(unique_sfs)
            # Verwende tab10 oder Set3 für gut unterscheidbare Farben
            if n_colors <= 10:
                colors = plt.cm.tab10(np.linspace(0, 1, 10))[:n_colors]
            else:
                colors = plt.cm.viridis(np.linspace(0, 1, n_colors))
            color_map = {sf: colors[i] for i, sf in enumerate(unique_sfs)}
            
            # RSSI Plot
            # Zuerst die Linie
            ax1.plot(data['timestamp'], data[rssi_col], 'k-', alpha=0.2, linewidth=1, zorder=1)
            
            # Dann die Punkte für jeden SF
            for sf in unique_sfs:
                mask = data['sf_numeric'] == sf
                sf_data = data[mask]
                ax1.scatter(sf_data['timestamp'], sf_data[rssi_col], 
                           color=color_map[sf], s=60, label=f'SF{sf}', 
                           edgecolors='black', linewidth=0.5, alpha=0.8, zorder=2)
            
            ax1.set_ylabel('RSSI (dBm)', fontsize=12)
            ax1.set_title(f'RSSI und SNR über Zeit - {file_name}', fontsize=14)
            ax1.grid(True, alpha=0.3)
            ax1.legend(title='Spreading Factor', bbox_to_anchor=(1.02, 1), loc='upper left', 
                      borderaxespad=0, frameon=True, fancybox=True, shadow=True)
            
            # SNR Plot
            # Zuerst die Linie
            ax2.plot(data['timestamp'], data[snr_col], 'k-', alpha=0.2, linewidth=1, zorder=1)
            
            # Dann die Punkte für jeden SF
            for sf in unique_sfs:
                mask = data['sf_numeric'] == sf
                sf_data = data[mask]
                ax2.scatter(sf_data['timestamp'], sf_data[snr_col], 
                           color=color_map[sf], s=60, label=f'SF{sf}', 
                           edgecolors='black', linewidth=0.5, alpha=0.8, zorder=2)
            
            ax2.set_xlabel('Zeit', fontsize=12)
            ax2.set_ylabel('SNR (dB)', fontsize=12)
            ax2.grid(True, alpha=0.3)
            ax2.legend(title='Spreading Factor', bbox_to_anchor=(1.02, 1), loc='upper left', 
                      borderaxespad=0, frameon=True, fancybox=True, shadow=True)
            
            # Formatiere x-Achse
            fig.autofmt_xdate()
            
            plt.tight_layout()
            plt.show()
            
            # Statistiken ausgeben
            print(f'\nStatistiken für {file_name}:')
            
            # Zeitstatistiken
            start_time = data['timestamp'].min()
            end_time = data['timestamp'].max()
            duration = end_time - start_time
            
            print(f'\nZeitbereich:')
            print(f'  Startzeit: {start_time.strftime("%Y-%m-%d %H:%M:%S")}')
            print(f'  Endzeit:   {end_time.strftime("%Y-%m-%d %H:%M:%S")}')
            print(f'  Laufzeit:  {duration}')
            print(f'  Laufzeit (Minuten): {duration.total_seconds() / 60:.2f}')
            print(f'  Laufzeit (Stunden): {duration.total_seconds() / 3600:.2f}')
            
            print(f'\nSignalstatistiken:')
            print(f'  RSSI - Min: {data[rssi_col].min():.2f}, Max: {data[rssi_col].max():.2f}, Mittelwert: {data[rssi_col].mean():.2f}')
            print(f'  SNR - Min: {data[snr_col].min():.2f}, Max: {data[snr_col].max():.2f}, Mittelwert: {data[snr_col].mean():.2f}')
            
            # SF Statistiken
            print('\nStatistiken nach Spreading Factor:')
            for sf in unique_sfs:
                mask = data['sf_numeric'] == sf
                sf_data = data[mask]
                print(f'  SF{sf}: {len(sf_data)} Messungen, '
                      f'RSSI Mittel: {sf_data[rssi_col].mean():.2f} dBm, '
                      f'SNR Mittel: {sf_data[snr_col].mean():.2f} dB')
            print('-' * 80)
            
        else:
            print(f'Warnung: Spaltennamen fehlen in {file_name}.')
            print(f'Erwartet: timestamp, rssi/rssi_dbm, snr/snr_db, spreading_factor')
            
    except Exception as e:
        print(f'Fehler beim Verarbeiten von {file_name}: {str(e)}')
```
